```python
#reading and Write Multiimage format using OpenCV
import cv2
import numpy as np  
import matplotlib.pyplot as plt
grayimages= cv2.imreadmulti('MyMultiPics.tiff', flags=cv2.IMREAD_GRAYSCALE) 
# Display the number of images read
print(f"Number of images read: {len(grayimages[1])}") 
# show images in the same row
plt.figure(figsize=(10, 5))
for i in range(len(grayimages[1])):
    plt.subplot(1, len(grayimages[1]), i + 1)
    plt.imshow(grayimages[1][i], cmap='gray')
    plt.title(f'Image {i + 1}')
    plt.axis('off')
plt.tight_layout()
plt.show()
```
